### Check whether number is divisible by 3 or 7 in detail

```python
num = int(input('Enter the positive number : '))

if(num%3==0) and (num%7==0):
    print(f'Entered Number {num} is divisible by 3 and 7')
elif(num%3==0):
    print(f'Entered Number{num} is divisible by 3')
elif(num%7==0):
    print(f'Entered Number {num} is divisible by 7')
else:
    print(f'Entered Number {num} is not divisible by 3 or 7')
```
